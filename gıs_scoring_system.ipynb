```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #bununla model
from tensorflow.keras.layers import Dense #bununla da katman oluturulur
#dense layer btn nronlarn outputlarn birbirine bal olduu katman eidi
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
#12 nronlu 4 katman ekledim

model.add(Dense(1))
#1 tane k katman koydum(bundan activation functiona gerek yok)

model.compile(optimizer="adam",loss="mse")
#optimizer olarak adam setim (en iyi performans veren)
#loss olarak yine mean square error (mse) kullandm
```

```python
model.fit(x=x_train, y=y_train,validation_data=(x_test,y_test),batch_size=250, epochs=300)
```

```python
kayip_verisi = pd.DataFrame(model.history.history)
```

```python
kayip_verisi.head()
```

```python

kayip_verisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahmin_dizisi = model.predict(x_test)
```

```python
mean_absolute_error(y_test,tahmin_dizisi)
```
